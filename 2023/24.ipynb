```python
def process_input(ttt):
    lines = ttt.split('\n')
    results = []
    for line in lines:
        P, V = line.split(" @ ")
        P = np.array(list(map(int, P.split(","))))
        V = np.array(list(map(int, V.split(","))))
        results.append((P, V))
        
    return results
```
