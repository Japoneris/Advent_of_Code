```python
with open("puzzles/02", "r") as fp:
    data = fp.read()[:-1]
    
```

```python
def parse_ids(stri):
    return list(map(lambda x: x.split("-"), stri.split(",")))
    
```

```python
label = "PROD"
IDs = parse_ids(data)
```

```python
label = "TEST"
IDs = parse_ids(data_test)
```

```python
def check_invalid_v2(r0, r1):
    v0, v1 = int(r0), int(r1)

    vx = v0
    
    lst_invalid = []
    while True:
        if vx > v1:
            # Stop here
            break

        rx = str(vx)
        l = len(rx)
        if l % 2 == 1:
            # Move r0 to the nearest "big" number
            vx = 10**(l)
            # Start the loop back
            continue

        # Else: l0 is mod2
        l2 = l//2
        rbis = rx[:l2]*2
        vbis = int(rbis)
        if (vbis <= v1) & (vbis >= v0):
            lst_invalid.append(vbis)

        vx = (int(rx[:l2])+1) * (10**l2)

    return lst_invalid
        
    
```

```python
tot = 0
for ra, rb in IDs:
    s = check_invalid_v2(ra, rb)
    tot += sum(s)
print("Part 1:", label, tot)
```

```python
def check_invalid_v3(r0, r1, d=2):
    v0, v1 = int(r0), int(r1)

    vx = v0
    
    lst_invalid = []
    while True:
        if vx > v1:
            # Stop here
            break

        rx = str(vx)
        l = len(rx)
        if l % d != 0:
            # Move r0 to the nearest "big" number
            vx = 10**(l)
            # Start the loop back
            continue

        # Else: l0 is mod2
        l2 = l//d
        rbis = rx[:l2]*d
        vbis = int("0" + rbis)
        if (vbis <= v1) & (vbis >= v0):
            lst_invalid.append(vbis)

        vx = (int("0" + rx[:l2])+1) * (10**(l-l2))

    return lst_invalid
        
    
```

```python
tot = 0
for ra, rb in IDs:
    print(ra, rb)
    lst = [] # List for set conversion, as duplicate can exist
    for d in range(2, 10):
        s = check_invalid_v3(ra, rb, d)
        lst.extend(s)
        if len(s) > 0:
            print("Depth:", d)
            print(s)
            
    tot += sum(set(lst))

    print()
print(tot)
```
