```python

with open("puzzles/01", "r") as fp:
    data = fp.read()[:-1]
```

# Part 1

```python
r = 50
cnt = 0
for row in data.split():
    k = row[0]
    v = int(row[1:])
    if k == "R":
        r += v
    else:
        r -= v

    if (r % 100) == 0:
        cnt += 1
        
```

```python
print("Part 1:", cnt)
```

```python
Right 

99 => 100 dd = 1
99 => 101 dd = 1
100 => 101 dd = 0
100 => 200 dd = 1
100 => 201 dd = 1

101 => 102
101 => 200
101 => 201

Down 

102 => 101 
101 => 100 dd = 0 !!
101 => 99 dd = 1

100 => 99  dd = 1 KO
100 => 0   dd = 1 OK
200 => 99 dd = 1 OK

```

```python
r = 50
cnt = 0
for row in data.split():
    k = row[0]
    v = int(row[1:])

    r_past = r
    
    if k == "R":
        r += v
        # dd is sufficient to count if 0 or not
    else:
        r -= v
        if ((r % 100) == 0) & ((v % 100) != 0):
            cnt += 1
        if (r_past % 100) == 0:
            cnt -= 1

    dd = abs(r//100 - r_past//100)
    
    cnt += dd
```

```python
print("Part 2:", cnt)
```
