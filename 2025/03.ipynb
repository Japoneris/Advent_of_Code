```python
with open("puzzles/03", "r") as fp:
    data = fp.read()[:-1]
```

```python
lines = data.splitlines()
```

# Part 1

```python
def find_joltage(line):
    """
    strategy:
    Find the maximal digit available.
    Then, search left for the next best, then search right.
    """
    digits = [int(i) for i in line]
    v = max(digits)
    i = digits.index(v)

    left = digits[:i]
    right = digits[i+1:]

    v1 = 0
    if len(left) > 0:
        v_left = max(left)
        v1 = v_left*10 + v

    
    v2 = 0
    if len(right) > 0:
        v_right = max(right)
        v2 = v*10 + v_right
        
    return max(v1, v2)
```

```python
tot = 0
for line in lines:
    c = find_joltage(line)
    tot += c

print("Part 1:", tot)
```

# Part 2

```python
def find_best_joltage(numbers, digits=12):
    if digits == 1:
        return [max(numbers)]
        
    sub = numbers[:-(digits-1)]
    m = max(sub)
    i = sub.index(m)
    return [m] + find_best_joltage(numbers[i+1:], digits-1)
```

```python
tot = 0
for line in lines:
    ans = find_best_joltage(line)
    c = int("".join(ans))
    tot += c

print("Part 2:", tot)

```
