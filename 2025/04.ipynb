```python
with open("puzzles/04", "r") as fp:
    data = fp.read()[:-1].split()

n0 = len(data)
n1 = len(data[0])

data = list(map(lambda row: [x for x in row], data))
```

# Part 1

```python
def count_around(i0, j0, data=data):
    tot = 0
    for dx in [-1, 0, 1]:
        i = i0 + dx
        if (i < 0) | (i >= n0):
            continue
        for dy in [-1, 0, 1]:
            j = j0 + dy
            if (j < 0) | (j >= n1):
                continue

            if data[i][j] == "@":
                tot += 1

    return tot - 1 # account for the middle
```

```python
tot = 0
for i in range(n0):
    for j in range(n1):
        if data[i][j] != "@":
            continue
        # Else
        c = count_around(i, j)
        if c < 4:
            tot += 1

print("Part 1:", tot)
```

# Part 2

```python
tot_full = 0

while True:
    
    tot = 0
    for i in range(n0):
        for j in range(n1):
            if data[i][j] != "@":
                continue
            # Else
            c = count_around(i, j, data)
            if c < 4:
                tot += 1
                
                data[i][j] = "."
    # Update
    tot_full += tot
    if tot == 0:
        break
    else:
        print(tot)
                

print("Part 2:", tot_full)
```
