```python
with open("puzzles/11", "r") as fp:
    data = fp.read()[:-1].split("\n")
```

```python
dico = {}
for line in data:
    key, items = line.split(": ")
    items = items.split(" ")

    dico[key] = items
```

# Part 1: Simple memoization trick

```python
# Need DFS to memoize the number of times the path reach the end or not
dic_memo = {"out": 1}

def explore(node):
    
    if node in dic_memo:
        return dic_memo[node]

    if node == "dac":
        print(node)
    elif node == "fft":
        print(node)

    # Else, explore and store
    tot = 0
    for item in dico[node]:
        tot += explore(item)

    dic_memo[node] = tot
    return tot
```

```python
print("Part 1:", explore("you"))
```

# Part 2: should visit `dac` and `fft` node to validate a path ...

Start from svr

svr -> dac: 469271495356 path

svr -> out: 65641226304821230 path
svr -> fft: 7920 path
fft -> dac: 6662060 path
dac -> fft: 0
dac -> out: 7314

Manually do the job.

```python
# Need DFS to memoize the number of times the path reach the end or not
dic_memo = {"out": 1}

def explore2(node, flag=0):
    
    if node in dic_memo:
        return dic_memo[node]

    # Else, explore and store
    tot = 0
    for item in dico[node]:
        tot += explore2(item, flag)

    dic_memo[node] = tot
    return tot

explore2("dac")
```

```python
result = 7920 * 6662060 * 7314
print("Part 2:", result)
```
