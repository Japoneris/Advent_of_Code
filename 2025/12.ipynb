```python
with open("puzzles/12", "r") as fp:
    data = fp.read()
```

```python
def parse_puzzle(data):
    *lst_present, lst_blocks = data.split("\n\n")

    lst_parsed_present = []
    for line in lst_present:
        _, b = line.split(":\n")
        lst_parsed_present.append(b.split("\n"))

    lst_parsed_blocks = []
    for line in lst_blocks.split("\n")[:-1]:
        area, counts = line.split(": ")
        counts = [int(x) for x in counts.split()]
        a, b = area.split("x")
        lst_parsed_blocks.append(((int(a), int(b)), counts))
        
        
    return lst_parsed_present, lst_parsed_blocks
```

```python
blocks, area_parts = parse_puzzle(data)
```

# Part 1

## Attempt 1: discard based on total area

```python
block_size = {}
for idx, block in enumerate(blocks):
    c = "".join(block).count("#")
    block_size[idx] = c
    
```

```python
lst_valid = []
cnt = 0
for area, block_count in area_parts:
    a, b = area
    area_tot = a * b 
    needed  = sum([block_size[i] * v for i, v in enumerate(block_count)])
    if needed > area_tot:
        #print(area_tot, needed)
        cnt += 1
    else:
        print("Diff: {}".format(area_tot - needed))
        lst_valid.append((area, block_count))
        
print("{} invalid parts. At best: {} / {}".format(cnt, len(area_parts) - cnt, len(area_parts)))
```
