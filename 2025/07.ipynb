```python
with open("puzzles/07", "r") as fp:
    data = fp.read().split()
```

# Part 1

```python
tot = 0

i0 = data[0].find("S")

matrix = [[x for x in row] for row in data[1:]]
matrix[0][i0] = "|"

for idx, line in enumerate(matrix[:-1]):
    lst = [j for j, v in enumerate(line) if v == "|"]
    next_line = matrix[idx+1]
    for p in lst:
        if next_line[p] == "^":
            matrix[idx+1][p-1] = "|"
            matrix[idx+1][p+1] = "|"
            tot += 1
        else:
            matrix[idx+1][p] = "|"
    
          
    
```

```python
print("Part 1:", tot)
```

# Part 2

```python
tot = 0

i0 = data[0].find("S")

dic_map = {".": 0, "^": "^"}

matrix = [[dic_map[x] for x in row] for row in data[1:]]
matrix[0][i0] = 1

for idx, line in enumerate(matrix[:-1]):
    lst = [(j, v) for j, v in enumerate(line) if isinstance(v, int) and v > 0]
    
    next_line = matrix[idx+1]
    for p, v in lst:
        if next_line[p] == "^":
            matrix[idx+1][p-1] += v
            matrix[idx+1][p+1] += v
            tot += 1
        else:
            matrix[idx+1][p] += v
    
          
    
```

```python
print("Part 2:", sum(matrix[-1]))
```
