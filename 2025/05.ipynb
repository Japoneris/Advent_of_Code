```python
with open("puzzles/05", "r") as fp:
    data = fp.read()[:-1].split()
```

```python
def extract_info(lst):
    lst_fresh = []
    lst_available = []
    for item in lst:
        if "-" in item:
            a, b = item.split("-")
            lst_fresh.append((int(a), int(b)))
        else:
            lst_available.append(int(item))

    return lst_fresh, lst_available
```

```python
fresh, available = extract_info(data)
```

```python
tot = 0
for x in available:
    for ra, rb in fresh:
        if (x >= ra) & (x <= rb):
            tot += 1
            break
            
```

```python
print("Part 1:", tot)
len(fresh), len(available)
```

# Part 2

```python
fresh, available = extract_info(data)

fresh = sorted(fresh, key=lambda x: x[0])
fresh = [[x[0], x[1]] for x in fresh] # Convert tuple to list
```

```python
tot = 0
n = len(fresh)
for i in range(n):
    ra, rb = fresh[i]
    if ra <= rb:
        tot += (rb-ra+1)
        #print(ra, rb, rb-ra+1)

    # Update all the sub items
    for j in range(i+1, n):
        fresh[j][0] = max(fresh[j][0], rb+1)

print("Part2:", tot)
```
