```python
with open("puzzles/06", "r") as fp:
    data = fp.read()
```

```python
*lines, maths = data.splitlines()
```

# Part 1

```python
values = [x.split() for x in lines]

ops = maths.split()
```

```python
from functools import reduce
```

```python
tot = 0
for idx, op in enumerate(ops):
    lst = [int(line[idx]) for line in values]
    if op == "*":
        tot += reduce(lambda x, y: x*y, lst)
    else:
        tot += sum(lst)
        
```

```python
print("Part 1:", tot)
```

```python
lst_verticals = [-1]
for i in range(len(lines[0])):
    flag = True
    for line in lines:
        if line[i] != " ":
            flag = False

    if flag:
        lst_verticals.append(i)

lst_verticals.append(len(lines[0]))
```

```python
tot = 0
for op, a, b in zip(ops, lst_verticals[:-1], lst_verticals[1:]):

    dx = b-a
    lst = []
    for i in range(1, dx):
        tmp = []
        for line in lines:
            c = line[a+i]
            if c != " ":
                tmp.append(c)
        lst.append(int("".join(tmp)))
    print(lst)
                   
    #print(line[a+1:b])
    if op == "*":
        tot += reduce(lambda x, y: x*y, lst)
    else:
        tot += sum(lst)
    

    print("=============")
```
