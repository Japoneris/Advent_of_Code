```python
with open("puzzles/08", "r") as fp:
    data = fp.read().split()
```

```python
def parse_values(line):
    a, b, c = line.split(",")
    return int(a), int(b), int(c)
               
```

```python
lines = list(map(parse_values, data))
```

```python
def dist(val0, val1):
    x0, y0, z0 = val0
    x1, y1, z1 = val1
    return (x0-x1)**2 + (y0-y1)**2 + (z0-z1)**2
    
lst_dist = []
for i, val_i in enumerate(lines):
    for j, val_j in enumerate(lines[:i]):
        d = dist(val_i, val_j)
        lst_dist.append((d, i, j))

# Order the groups
lst_dist = sorted(lst_dist, key=lambda x: x[0])
```

##### Cluster

```python
l = len(lines)
dic_ID_group = dict([(idx, idx) for idx in range(l)])
dic_group_IDs = dict([(idx, [idx]) for idx in range(l)])
```

### Aggregate

```python
for d, i, j in lst_dist[:1000]:
    gi = dic_ID_group[i]
    gj = dic_ID_group[j]
    
    if gi == gj:
        # Already in same group, skip
        continue
    else:
        lst_j = dic_group_IDs[gj]
        for v in lst_j:
            dic_ID_group[v] = gi
            dic_group_IDs[gi].append(v)
        del(dic_group_IDs[gj])
```

#### Count group size

```python
lst_size = sorted(map(len, dic_group_IDs.values()), reverse=True)
```

```python
tot = 1
for key in list(lst_size)[:3]:
    tot = tot * key

print("Part 1:", tot)
```
